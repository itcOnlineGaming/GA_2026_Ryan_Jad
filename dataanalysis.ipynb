```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

df = pd.read_csv('dataanalysis.csv')

plt.figure()
plt.scatter(df["Speed"], df["Survival_Time_Seconds"])
plt.xlabel("Player Speed")
plt.ylabel("Time Survived")
plt.title("Time Survived vs Player Speed")
plt.show()
avg_scores = df.groupby("Speed")["Survival_Time_Seconds"].mean()
```
